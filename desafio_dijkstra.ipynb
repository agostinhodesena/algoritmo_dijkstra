```python
def pedir_vertice(mensagem, grafo):

    while True:
        v = input(mensagem).strip()

        if v in grafo:
            return v
        else:
            print(f"Vértice {v} não existe no grafo. Tente novamente!")

def definir_parametro():
    
    parametro = int(input("Digite 1 para que o parâmetro seja distância ou 0 para que seja tempo:"))

    if parametro == 1:
        parametro_definido = True
    else:
        parametro_definido = False
    
    return parametro_definido
```

```python
ponto_de_partida = 'Rua Augusta, 500'

def obter_menor_valor(parametro_definido, arestas):

    if parametro_definido:
        menor_vizinho = min(
            arestas, 
            key = lambda v: arestas[v]['distancia']
            )
        menor_valor = arestas[menor_vizinho]['distancia']
    else:
        menor_vizinho = min(
            arestas, 
            key = lambda v: arestas[v]['tempo']
            )
        menor_valor = arestas[menor_vizinho]['tempo']
    
    return menor_vizinho, menor_valor
      

def definindo_caminhos(grafo, parametro_definido, ponto_de_partida, ponto_de_chegada):

    valores = {vertice: float('inf') for vertice in grafo} # Cada vertice do mapa tem como distância/tempo o infinito porque não conhecemos ela.
    vertices_anteriores = {vertice: None for vertice in grafo} # Como não sabemos qual é o vertice anterior no caminho proposto, dizemos que o vértice anterior é a ausência de um valor.
    valores[ponto_de_partida] = 0 # Todo ponto de partida começa com valor de distancia 0, já que ele está no início
    vertices_visitados = set() # É criado uma coleção zerada para armazenar todos os vertices visitados

    while True:
        # Definindo o vértice não visitado com menor valor de distancia/tempo:
        vertice_atual = min((v for v in grafo if v not in vertices_visitados),
                            key=lambda v: valores[v], default=None)
        
        # Se o vértice atual for ausência de valor ou se não houver mudança de valor:
        if vertice_atual is None or valores[vertice_atual] == float('inf'):
            break

        # Se o vértice atual for o ponto de chegada:
        if vertice_atual == ponto_de_chegada:
            break
        
        # Adicionando o vertice atual ao conjunto de vertices visitados:
        vertices_visitados.add(vertice_atual)

        # Cada vértice tem um conjunto de arestas, ou seja, um conjunto de caminhos.
        # No vértice atual, vamos:
        for vertice, arestas in grafo[vertice_atual].items():
            # Definir qual parêmtro queremos...
            peso = arestas['distancia'] if parametro_definido else arestas['tempo']
            # Somar ao infinito a distância proposta pelo desafio:
            valor_atualizado = valores[vertice_atual] + peso
            
            # Sempre o infinito vai ser maior que o valores do vértice exceto no caso em que:
            # 1) Você for o ponto inicial, pois será 0
            if valor_atualizado < valores[vertice]:
                valores[vertice] = valor_atualizado
                vertices_anteriores[vertice] = vertice_atual

    caminho_final = []
    vertice_final = ponto_de_chegada

    while vertice_final is not None:
        caminho_final.insert(0, vertice_final)
        vertice_final = vertices_anteriores[vertice_final]

    return caminho_final, valores[ponto_de_chegada]
```

```python
ponto_de_partida = pedir_vertice('Digite o ponto de partida:', grafo)
ponto_de_chegada = pedir_vertice('Digite o ponto de chegada:', grafo)
parametro_definido = definir_parametro()

print(f"Ponto de Partida: {ponto_de_partida}")
print(f"Ponto de Chegada: {ponto_de_chegada}")
print(f"Parâmetro definido: {"Distância" if parametro_definido is True else "Tempo"}")

caminho_final, valor_total = definindo_caminhos(grafo, parametro_definido, ponto_de_partida, ponto_de_chegada)
```
